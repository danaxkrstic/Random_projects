```python
def Rastav(e):
    e=bin(e)[2:]
    e_rastav=list(e)
    temp=[]
    for a in e_rastav:
      temp.append(int(a))
    temp=temp[::-1] #obrnemo poredak
    return temp
    
```

```python
def Prost(x): 
  if x > 1:
      for i in range(2, int(x/2)+1):
        
          if (x % i) == 0:
              return False
      else:
          return True
  else:
      return False

```

```python
def modInverse(A, M):

    for X in range(1, M):
        if (((A % M) * (X % M)) % M == 1):
            return X
    return -1
```

```python
def ASCII(lista): #prevorba u ascii
  temp=[]
  for a in lista:
    temp.append(ord(a))

  return temp
```

```python
def square_and_multiply_always(x,n,bin):
  start_time = time.time()
  y=1

  for i in range(len(bin)-1,-1,-1):
    y=y**2%n
    if(bin[i]==0):
      dummy=(y*x)%n
    else:
      y=(y*x)%n

  return (y,time.time() - start_time)
```

```python
def Montgomery_ladder(x,n,bin):
  start_time = time.time()
  y0=1
  y1=x
  for i in range(len(bin)-1,-1,-1):
    if bin[i]==0:
      temp=y0
      y0=(y0**2)%n
      y1=(y1*temp)%n
    else:
      y0=(y0*y1)%n
      y1=(y1**2)%n

  return (y0,time.time() - start_time)
```

```python
def Multiplicative_splitting_modulo_N(x,n):
  a=[0,1]
  y=[n,x]
  i=1

  q=["x"]

  while y[i]>=math.ceil(math.sqrt(n)):
    q.append("x") # da ne bude index out of range
    y.append("x")
    a.append("x")

    q[i]=(y[i-1]//y[i])%n
    y[i+1]=(y[i-1]-q[i]*y[i])%n
    a[i+1]=(a[i-1]-q[i]*a[i])%n
    i+=1
  
  return a[i],y[i]
```

```python
def Regular_exp_with_halfsize_multiplic(x,n,bin):
  start_time = time.time()
  x_0,x_1=Multiplicative_splitting_modulo_N(x,n)
  y= modInverse(x_0,n) #modularni inverz
  for i in range(len(bin)-1,-1,-1):
    y=(y**2)%n
    if bin[i]==0:
      y=(y*x_0)%n
    else:
      y=(y*x_1)%n

  y=(y*x_0)%n


  return (y,time.time() - start_time)
```
